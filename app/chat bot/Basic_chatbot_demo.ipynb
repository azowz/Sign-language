Let’s build a chatbot that answers your questions using the **OpenAI API**!

```python
import os
from dotenv import load_dotenv

# Load env vars from .env file
load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
print("Key loaded:", bool(OPENAI_API_KEY))  # Should be True
```

Chatbot Class and Basic Test

**1. Interactive Loop for Asking Multiple Questions**
Now let’s make the chatbot interactive! This code lets you ask multiple questions and exit by typing “exit”. The AI answers each question, and the conversation is saved.

Add loop logic. Run code, enter your API key, and try asking a few questions!

```python
import openai

class Chatbot:
    def __init__(self, api_key: str, model: str = "gpt-3.5-turbo"):
        """Initialize the chatbot with an API key and model."""
        self.api_key = api_key
        self.client = openai.OpenAI(api_key=api_key)  # Create OpenAI client
        self.model = model
        self.history = []  # Store conversation history

    def ask_question(self, question: str) -> str:
        """Send a question to the AI and get an answer."""
        # Add user question to history
        self.history.append({"role": "user", "content": question})

        # Send request to OpenAI API
        response = self.client.chat.completions.create(
            model=self.model,
            messages=self.history
        )

        # Extract assistant's answer
        ai_answer = response.choices[0].message.content

        # Save assistant answer in history
        self.history.append({"role": "assistant", "content": ai_answer})

        return ai_answer

    def show_history(self) -> list:
        """Show all questions and answers in the conversation."""
        return self.history



api_key = OPENAI_API_KEY

bot = Chatbot(api_key)

# Interactive loop to ask multiple questions from the chatbot
count_question = 0   # Counter for the number of questions asked
max_questions = 100  # Limit to avoid infinite loop

while count_question < max_questions:  
    # Ask the user for input
    question = input("What is your question? (type 'stop' to finish): ")

    # Check if user wants to stop
    if question.lower() == "stop":
        print("Conversation ended.")
        break

    # Send question to chatbot and print answer
    print("AI answers:", bot.ask_question(question))

    # Increase counter after each question
    count_question += 1

    # Optional: show full conversation so far
    print("\n--- Conversation so far ---")
    print(bot.show_history())
    print("---------------------------\n")
```
